```python
result = results.split('\n')[:-2][0]
result
```

```python
'名詞' in result.split('\t')[4]
# 品詞の情報は前から数えて5番目の要素のため、配列のインデックス4で指定する
# 名詞のみを抽出する場合、この文字列の中に名詞が含まれるかを判定する処理を入れる
```

```python
for result in results.split('\n')[:-2]:
    if '名詞' in result.split('\t')[4]:
        print(result.split('\t')[0])
# 名詞だったら、実際の単語をprintする処理、名詞のみを抽出
```
